```python
bsize = 10
dlen = len(angles_yedek)
splitpoint = int(0.75*dlen)
reindex = list(range(len(angles_yedek)))
random.seed(1234)
random.shuffle(reindex)
len(images_yedek), len(angles_yedek), len(yonler_yedek)
```

```python
def augment_brightness(image):
    image1 = cv2.cvtColor(image,cv2.COLOR_BGR2HSV) 
    image1 = np.array(image1, dtype = np.float64)
    random_bright = .5+np.random.uniform()
    image1[:,:,2] = image1[:,:,2]*random_bright
    image1[:,:,2][image1[:,:,2]>255]  = 255
    image1 = np.array(image1, dtype = np.uint8)
    image1 = cv2.cvtColor(image1,cv2.COLOR_HSV2RGB)
    return image1

def get_matrix(fname):
    img = cv2.imread(fname)
    if random.randint(0,1) == 1 :
        img = augment_brightness(img)        
    return img[350:600,200:1200,:]

def get_yon(fname):
    yon_img = cv2.imread(fname + '.jpeg', 0)
    yon_img = cv2.resize(yon_img, (50,50))
    return yon_img[:,:]

def generate_data():
    i = 0
    while True:
        x_1 = []
        x_2 = []
        y = []
        for j in range(i,i+bsize): 
            ix = reindex[j]
            x_1.append(get_matrix(images_yedek[ix]))
            x_2.append(get_yon(yonler_yedek[ix]))
            y.append(np.array([angles_yedek[ix]]))
        x_1 = np.array(x_1)
        x_2 = np.array(x_2)
        y = np.array(y)
        yield ([x_1, x_2],y)
        i +=bsize
        if i+bsize > splitpoint:
            i = 0

# Generate data for validation                  
def generate_data_val():
    i = splitpoint
    while True:
        x_1 = []
        x_2 = []
        y = []
        for j in range(i,i+bsize): 
            ix = reindex[j]
            x_1.append(get_matrix(images_yedek[ix]))
            x_2.append(get_yon(yonler_yedek[ix]))
            y.append(np.array([angles_yedek[ix]]))
        x_1 = np.array(x_1)
        x_2 = np.array(x_2)
        y = np.array(y)
        yield ([x_1, x_2],y)
        i +=bsize
        if i+bsize > dlen:
            i = splitpoint
```

```python
from tensorflow.keras import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Lambda, Layer, Concatenate
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Cropping2D, Activation
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import SGD
```

```python
model1 = Sequential()
model1.add(Lambda(lambda x: x / 127.5 - 1.0, input_shape=(250, 1000, 3)))
model1.add(Conv2D(24, (5, 5), activation="relu", strides=(2, 2)))
model1.add(Conv2D(36, (5, 5), activation="relu", strides=(2, 2)))
model1.add(Conv2D(48, (3, 3), activation="relu", strides=(2, 2)))
model1.add(Conv2D(64, (3, 3), activation="relu", strides=(1, 1)))
model1.add(Flatten())

model2 = Sequential()
model2.add(Lambda(lambda x: x/ 127.5 - 1.0, input_shape=(50, 50, 1)))
model2.add(Conv2D(24, (5, 5), activation="relu", strides=(2, 2)))
model2.add(Conv2D(36, (5, 5), activation="relu", strides=(2, 2)))
model2.add(Conv2D(48, (3, 3), activation="relu", strides=(2, 2)))
model2.add(Flatten())

model3 = Concatenate()([model1.output, model2.output])
model3 = Dense(100)(model3)
model3 = Dense(50)(model3)
model3 = Dense(10)(model3)
model3 = Dense(1)(model3)

model4 = Model([model1.input, model2.input], model3)
model4.compile(loss='mse', optimizer='adam')
```

```python
model_checkpoint = ModelCheckpoint(weights_path + 'yonlu_dnm.h5', monitor='val_loss', save_best_only=True)
```

```python
hs = model4.fit(generate_data(),steps_per_epoch=int(splitpoint/ bsize),
                    validation_data=generate_data_val(), 
                    validation_steps=(dlen-splitpoint)/bsize, epochs=10,callbacks=[model_checkpoint])
```
